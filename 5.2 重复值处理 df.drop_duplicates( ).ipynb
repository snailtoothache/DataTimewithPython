```python
import pandas as pd
import numpy as np
```

```python
# drop_duplicates() 对所有值进行重复判断，且默认保留第一个（行）值
df = pd.DataFrame ([['A1','张通',101,'2018-08-08'],
                   ['A2','李谷',102,'2018-08-09'],
                   ['A3','孙凤',103,'2018-08-10'],
                   ['A3','孙凤',103,'2018-08-10'],
                   ['A4','赵恒',104,'2018-08-11'],
                   ['A5','赵恒',104,'2018-08-12']],
                  columns = ('订单编号','客户姓名','唯一识别码','成交时间'))
df
```

```python
df.drop_duplicates()
```

```python
# 仅针对相关列作删除重复值操作
df
```

```python
df.drop_duplicates(subset = '唯一识别码')
```

```python
# 针对多列作删除重复值操作 
df
```

```python
df.drop_duplicates(subset = ['客户姓名','唯一识别码'])
```

```python
# 保留最后一个重复值
# 传入keep 参数
df
```

```python
df.drop_duplicates(subset = ['客户姓名','唯一识别码'], keep='last')
```

```python
# 不保留任何重复值
# 设置False 把重复值全部删除
df
```

```python
df.drop_duplicates(subset=['客户姓名','唯一识别码'],keep = False)
```
