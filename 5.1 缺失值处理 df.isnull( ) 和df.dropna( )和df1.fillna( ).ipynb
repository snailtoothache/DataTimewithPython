```python
import pandas as pd
import numpy as np
```

```python
# 5.1.1 缺失值查看
df = pd.DataFrame ([['A1',54,'男','2018/8/8'],['A2',16,np.nan,'2018/8/9'],['A3',47,'女','2018/8/10'],['A4',41,'男','2018/11']],columns = ['编号','年龄','性别','注册时间'])
df
```

```python
df.info()
```

```python
# isnull()方法判断哪个值是缺失值？
df
```

```python
df.isnull()
```

```python
# 5.1.2 缺失值删除 （缺失某个字段/空白行）
# dropna()方法,删除含有缺失值的一行
df
```

```python
df.dropna()
```

```python
# 仅删除全部空值的行
df1 = pd.DataFrame ([['A1',54,'男','2018/8/8'],['A2',16,np.nan,'2018/8/9'],[np.nan,np.nan,np.nan,np.nan],['A4',41,'男','2018/11']],columns = ['编号','年龄','性别','注册时间'])
df1
```

```python
df1.dropna(how = 'all')
```

```python
# 5.1.3 缺失值填充
# fillna() 填充
df1
```

```python
df1.fillna(0)
```

```python
# 不同列的fillna() 填充
df2 = pd.DataFrame ([['A1',54,'男','2018/8/8'],['A2',16,np.nan,'2018/8/9'],['A3',np.nan,'女','2018/8/10'],['A4',41,'男','2018/11']],columns = ['编号','年龄','性别','注册时间'])
df2
```

```python
# 对性别进行填充
df2.fillna({'性别':'男'})
```

```python
# 不同列的fillna() 填充
df2
```

```python
# 对年龄进行填充
df2.fillna({"年龄":"30.0"})
```

```python
# 对年龄、性别进行填充
df2.fillna({'年龄':'30.0', '性别':'男'})
```
