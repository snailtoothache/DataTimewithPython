```python
# 14.3.1 哪些类别的商品比较畅销
# 要看哪些类别的商品比较畅销，只要将订单表中的数据按照类别ID进行分组，然后对分组后的销量求和，就会得到每一类在一段时间内的销量
data.groupby('类别ID')['销量'].sum().reset_index()
```

```python
# 获取所有类别在一段时间内对应的销量，查看销量最好的前10个类别，对销量做一个降序排序，然后取前10行即可
data.groupby('类别ID')['销量'].sum().reset_index().sort_values(by='销量',ascending=False).head(10)
```

```python
# 14.3.2 哪些商品比较畅销
# 计算哪些商品比较畅销，利用数据透视表来计算哪些商品比较畅销，同样取前10名的商品
pd.pivot_table(data,index='商品ID',values='销量',aggfunc='sum').reset_index().sort_values(by='销量',ascending=False).head(10)
```

```python
# 14.3.3 不同门店的销售额占比
# 新增一个销售额字段
# 按照门店编号进行分组，对分组后的营业额求和运算，最后计算不同门店的销售额占比
```

```python
data['销售额'] = data['销量']*data['单价']
data.groupby('门店编号')['销售额'].sum()
```

```python
data.groupby('门店编号')['销售额'].sum()/data['销售额'].sum()
```

```python
# 绘制饼图
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
import numpy as np
(data.groupby('门店编号')['销售额'].sum()/data['销售额'].sum()).plot.pie()
```

```python
# 14.3.4 哪些时间段是超市的客流高峰期
# 了解清楚哪些时间是超市客流的高峰期是很有必要的
# 可以帮助超市管理人员提前布置工作人员，帮助超市管理人员决定在什么时间段开展促销过的
# 查看一天中什么时间（哪几个小时）是高峰期，需要知道每个时间段对应的客流量，但是订单表中的成交时间既有日期又有时间，我们需要从中提取小时数
# 这里依然使用订单ID去重计数代表客流量
```

```python
# 利用自定义时间格式函数strftime提取小时数
data['小时']=data['成交时间'].map(lambda x:int(x.strftime('%H')))
# 对小时和订单去重
traffic=data[['小时','订单ID']].drop_duplicates()
# 求每小时的客流量
traffic.groupby('小时')['订单ID'].count()
```

```python
# 绘制每个小时客流量折线图
traffic.groupby('小时')['订单ID'].count().plot()
```

```python
# 分小时客流量折线图所示：可以看出8点到10点是超市一天中的销售高峰期，17点到19点又有一个销售小高峰
```
