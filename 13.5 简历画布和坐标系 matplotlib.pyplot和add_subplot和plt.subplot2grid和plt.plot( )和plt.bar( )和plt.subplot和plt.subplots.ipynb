```python
# 生成2行2列，且在（0，0）位置绘图的坐标系
plt.subplot2grid((2,2),(0,0))
```

```python
import numpy as np
x = np.arange(6)
y = np.arange(6)
```

```python
# 将图表的整个区域分为2行2列，且在（0，0）位置做折线图
plt.subplot2grid((2,2),(0,0))
plt.plot(x,y)      
```

```python
# 将图表的整个区域分为2行2列，且在（0，1）位置做柱形图
```

```python
plt.subplot2grid((2,2),(0,1))
plt.bar(x,y)
```

```python
# 13.5.4 用plt.subplot函数建立坐标系
# plt.subplot和plt.subplot2grid同属于plt库的函数，两者的区别只是表现形式不一样
plt.subplot(2,2,1)
```

```python
import numpy as np
x=np.arange(6)
y=np.arange(6)
```

```python
# 将图表的整个区域分为2行2列，且在第1个坐标系上做折线图
plt.subplot(2,2,1)
plt.plot(x,y)
```

```python
# 将图表的整个区域分为2行2列，且在第4个坐标系上做柱形图
plt.subplot(2,2,4)
plt.bar(x,y)
```

```python
# 13.5.5 用plt.subplots函数建立坐标系
# plt.subplots是plt库的一个函数，与前两个不同之处在于，subplot2grid和subplot函数每次只返回一个坐标系
# subplots函数一次可以返回多个坐标系
```

```python
# 将整个图表分成2行2列，并将4个坐标系全部返回
fig,axes=plt.subplots(2,2)
```

```python
# 通过axes[x,y]指明哪个坐标系里绘图
import numpy as np
x=np.arange(6)
y=np.arange(6)
```

```python
# 在[0,0]坐标系中绘制折线图
fig,axes=plt.subplots(2,2)
axes[0,0].plot(x,y)
```

```python
# 在[1,1]坐标系中绘制柱状图
fig,axes=plt.subplots(2,2)
axes[1,1].bar(x,y)
```

```python
# 13.5.6 几种创建坐标系方法的区别
# add_subplot属于对象式编程，所有的操作都是针对某个对象进行的，比如先建立一块画布，然后在这块画布上建立坐标系，进而在坐标系上绘图
# 后三种属于函数式编程，都是直接调用plt库里面的某个函数或者方法达到创建坐标系的目的
# 对象是编程的代码比较繁琐，但是便于理解
# 函数式编程虽然代码简洁，但是不利于新人掌握整体的绘图原理
# 这两种编程方式不仅体现在创建坐标系中，在接下来的一些操作也会有涉及，有的时候两者会交叉使用
```
