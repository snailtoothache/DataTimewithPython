```python
# 13.6.1 设置坐标轴的标题
# 横轴的标题为月份，纵轴的标题为注册人数
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('月份')
plt.ylabel('注册量')
```

```python
# 还可设置xlabel和ylabel到x轴和y轴的距离，给参数labelpad传入具体的距离数即可
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('月份',labelpad=10)
plt.ylabel('注册量',labelpad=10)
```

```python
# 还可以对xlabel和ylabel的文本相关性质进行设置，比如设置字体大小、字体颜色、是否加粗等
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('月份',fontsize='xx-large',color='#70AD47',fontweight='bold')
plt.ylabel('注册量')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
import numpy as np
```

```python
# 把图表中的x轴刻度值均定义成月份，y轴的刻度值均定义注册人数
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('月份',fontsize='xx-large',color='#70AD47',fontweight='bold')
plt.ylabel('注册人数')
# 设置x轴刻度
plt.xticks(np.arange(9),['1月份','2月份','3月份','4月份','5月份','6月份','7月份','8月份','9月份'])
# 设置y轴刻度
plt.yticks(np.arange(1000,7000,1000),['1000人','2000人','3000人','4000人','5000人','6000人'])
```

```python
# 将xticks、yticks传入一个空列表，可把x/y轴的数值隐藏
# 把图表中的x轴刻度值均定义成月份，y轴的刻度值均定义注册人数
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('月份',fontsize='xx-large',color='#70AD47',fontweight='bold')
plt.ylabel('注册人数')
plt.xticks([])
plt.yticks([])
```

```python
# 除了xticks、yticks方法，还可以使用plt库中tick_params函数对轴刻度线进行设置
# （此处仅记录，不运行代码）
plt.tick_params(axis,reset,which,direction,length,width,
                color,pad,labelsize, label_color,bottom,
                top,left,right,labelbottom,labeltop,labelleft,labelright)
```

```python
# tick_params函数中的参数及说明如下表所示
df = pd.DataFrame([['axis','对哪个轴的刻度线进行设置，x、y、both三个可选'],
                   ['reset','是否重置所有设置，True/False'],
                   ['which','对哪种刻度进行设置，major(主刻度线)、minor(次刻度线)、both三个可选'],
                   ['direction','刻度线的朝向，in(朝里)、out(朝外)、inout(里外均有)三个可选'],
                   ['length','刻度线长度'],
                   ['width','刻度线宽度'],
                   ['color','刻度线颜色'],
                   ['pad','刻度线与刻度标签之间的距离'],
                   ['labelsieze','刻度标签大小'],
                   ['labelcolor','刻度标签颜色'],
                   ['top、bottom、left、right','True/False可选，控制上、下、左、右刻度线是否显示'],
                   ['labeltop、Labelbottom、labelleft、labelright','True/False可选，控制上、下、左、右刻度标签是否显示'],
                   ['labelcolor','刻度标签颜色']],
                 columns=('参数','说明'))
df
```
