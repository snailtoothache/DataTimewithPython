```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
```

```python
# 9.4 时间索引
# 时间索引就是根据时间来对时间格式的字段进行数据选取的一种索引方式
```

```python
# 新建一个时间索引的DataFrame
index = pd.DatetimeIndex(['2018-01-01',
                          '2018-01-02',
                          '2018-01-03',
                          '2018-01-04',
                          '2018-01-05',
                          '2018-01-06',
                          '2018-01-07',
                          '2018-01-08',
                          '2018-01-09',
                          '2018-01-10'])
```

```python
data=pd.DataFrame(np.arange(1,11),columns=['num'],index=index)
data
```

```python
# 获取2018年的数据
data['2018']
```

```python
# 获取2018年1月的数据
data['2018-01']
```

```python
# 获取2018年1月1日到2018年1月5日的数据
data['2018-01-01':'2018-01-05']
```

```python
# 获取2018年1月1日的数据
data['2018-01-01':'2018-01-01']
```

```python
# 利用前面学过的布尔索引来对非索引列的时间进行选取
from dateutil.parser import parse
df = pd.DataFrame ([['张通',101,31,parse("2018-08-08")],
                   ['李谷',102,45,parse("2018-08-09")],
                   ['孙凤',103,23,parse("2018-08-10")],
                   ['赵恒',104,36,parse("2018-08-11")],
                   ['王娜',105,21,parse("2018-08-11")]],
                  columns = ("客户姓名","唯一识别码","年龄","成交时间"),index = ("A1","A2","A3","A4","A5"))
df
```

```python
# 选取成交时间为2018年8月8日的订单
df[df["成交时间"] == datetime(2018,8,8)]
```

```python
# 选取成交时间在2018年8月9日之后的订单
df[df["成交时间"] > datetime(2018,8,9)]
```

```python
# 选取成交时间在2018年8月10日之前的订单
df[df['成交时间']<datetime(2018,8,10)]
```

```python
# 选取成交时间在2018年8月8日与2018年8月11日之间的订单
df[(df['成交时间']>datetime(2018,8,8))&(df['成交时间']<datetime(2018,8,11))]
```
