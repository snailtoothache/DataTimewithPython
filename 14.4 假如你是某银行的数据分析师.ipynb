```python
# 14.4.1 是不是收入越高的人坏账率越低
# 先处理缺失值，然后选择均值填充方法
# 月收入列存在缺失值
```

```python
data=data.fillna({'月收入':data['月收入'].mean()})
data.info()
```

```python
# 月收入属于连续值，将连续值离散化，就是将连续值进行区间切分，分成若干类别
cut_bins=[0,5000,10000,15000,20000,100000]
income_cut=pd.cut(data['月收入'],cut_bins)
income_cut
```

```python
# 区间切分好以后，就可以看每个区间内的坏账率
# 坏账率就是所有借款用户中预期不还用户的占比
# 逾期不还用户的好坏客户字段标记为1
# 非逾期不还用户的好坏客户字段标记为0
# 坏账率=坏账客户数/所有借款用户的比值
all_income_user=data['好坏客户'].groupby(income_cut).count()
bad_income_user=data['好坏客户'].groupby(income_cut).sum()
bad_rate=bad_income_user/all_income_user
bad_rate
```

```python
# 绘制月收入与坏账率关系图
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
import numpy as np
bad_rate.plot.bar()
```

```python
# 当月收入在1万元以下时，收入越高，坏账率越低，当月收入超过1.5万元时，坏账率出现了上涨
```

```python
# 14.4.2 年龄和坏账率有什么关系
# 是否年龄越大消费越低呢？
```

```python
# 年龄也是连续值，也是用连续值离散化方式处理
age_cut=pd.qcut(data['年龄'],6)
all_age_user=data['好坏客户'].groupby(age_cut).count()
bad_age_user=data['好坏客户'].groupby(age_cut).sum()
bad_rate1=bad_age_user/all_age_user
bad_rate1
```

```python
# 绘制年龄好坏与坏账率关系图
bad_rate1.plot.bar()
```

```python
# 18岁以下的人的坏账率为0，18～36岁的人的坏账率最高
# 超高36岁的人随着年龄的增加，坏账率呈下降趋势
```

```python
# 14.4.3 家庭人口数量和坏账率有什么关系
```

```python
# 是否家庭人口越多，负担越重，坏账率越高？
```

```python
all_age_user=data.groupby('家属数量')['好坏客户'].count()
bad_age_user=data.groupby('家属数量')['好坏客户'].sum()
bad_rate2=bad_age_user/all_age_user
bad_rate2
```

```python
# 绘制家属数量与坏账率关系图
bad_rate2.plot()
```

```python
# 家属数量越多，负担越重，坏账率越高
# 当家属数量大于8人时，反而坏账率变为0
# 由于正常家庭人口数量不会有8个人，因此这部分数据可以当作异常值对待，删除不看即可
```
