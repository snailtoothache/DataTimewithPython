```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format='svg'
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
```

```python
# 13.7.1 网格线设置
# 网格线是相比于坐标轴更小的单位，网格线默认是关闭的
# 通过修改参数b的值=True来启动网格线
# 参数b=True,默认是将x轴和y轴的网格线全部打开
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.grid(b='True')
```

```python
# 修改axis的值控制打开哪个轴的网格线
# 只打开x轴的网格线
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.grid(b='True',axis='x')
```

```python
# 修改axis的值控制打开哪个轴的网格线
# 只打开y轴的网格线
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.grid(b='True',axis='y')
```

```python
# 网格线设置显示x轴或y轴，对网格线的线本身进行设置，比如线宽、线型、线的颜色的等
# 把网格线的线型（linestyle）设置成虚线（dashed),线宽（linewidth)设置为1
# 线型设置成虚线，线宽设置成1
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.grid(b='True',linestyle='dashed',linewidth=1)
```

```python
# 13.7.2 设置图例
# 图例对图表起到注释作用
# 在绘图的时候通过给label参数传入值表示该图表的图例名，再通过plt.legend()方法将图例显示出来
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.plot(x,y,label='折线图')
plt.bar(x,y,label='柱形图')
plt.legend()
```

```python
# 设置数据标签示例
# 在（5，1605）处显示该点的y值
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.title(label='1-9月XXX公司注册用户数',loc='center')
plt.text(5,1605,1605)
```

```python
# plt.text函数只是对针对坐标轴中的具体某一点（x,y)显示数值str
# 对整个图表显示数据标签，需要利用for进行遍历
# 在（x,y)处显示y值
x=[1,2,3,4,5,6,7,8,9]
y=[1000,2500,5500,6200,6000,1500,3500,4000,4200]
plt.subplot(1,1,1)
plt.plot(x,y)
plt.title(label='1-9月XXX公司注册用户数',loc='center')
for a,b in zip(x,y):plt.text(a,b,b,ha='center',va='bottom',fontsize=11)
```
